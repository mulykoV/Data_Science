```python
# Part 1
# NumPy
import numpy as np

# CRUD
# 1. Create
linear_data = np.arange(10)  # лінійні дані
random_data = np.random.random(10)  # випадкові дані

# 2. READ
print("Linear data: ", linear_data)
print("Random data: ", random_data)

# 3. Update
# Введення індексу для оновлення
choose = int(input("Enter index number (from 0 to 9) => "))

# Перевірка, чи індекс знаходиться в діапазоні
if 0 <= choose <= 9:
    linear_data[choose] = float(input("Enter number for linear data => "))
    random_data[choose] = float(input("Enter number for random data => "))
    print("\nLinear Data (updated):", linear_data)
    print("Random Data (updated):", random_data)
else:
    print("Invalid index. Please enter a number from 0 to 9.")

# 4. Delete
# Введення індексу для видалення
deleteNumber = int(input("Enter the index to delete (from 0 to 9) => "))

# Перевірка, чи індекс знаходиться в діапазоні
if 0 <= deleteNumber < len(linear_data):
    linear_data = np.delete(linear_data, deleteNumber)
    random_data = np.delete(random_data, deleteNumber)
    print("\nLinear Data (after delete):", linear_data)
    print("Random Data (after delete):", random_data)
else:
    print("Invalid index. Please enter a valid index.")
```

```python
# Part 2
# Імпорт бібліотеки Pandas
import pandas as pd
import numpy as np

# 1. Створення DataFrame різними способами

# 1.1 Створення DataFrame з масиву NumPy
data_np = np.random.rand(5, 3)  
df_numpy = pd.DataFrame(data_np, columns=['A', 'B', 'C'])
print("DataFrame з NumPy масиву:\n", df_numpy)

# 1.2 Створення DataFrame з використанням словників
data_dict = {
    'Name': ['Rose', 'Rododendron', 'Orchid', 'Tulip', 'Dahlia'],
    'Months of flowering': [3, 1, 5, 5, 12],
    'Spread': ['Holland', 'Botanical park', 'Home', 'Netherlands', 'Ukraine']
}
df_dict = pd.DataFrame(data_dict)
print("\nDataFrame зі словника:\n", df_dict)

# 2. Маніпуляції з даними

# 2.1 Відображення перших рядків (head)
print("\nПерші 3 рядки DataFrame зі словника:")
print(df_dict.head(3))

# 2.2 Опис статистики (describe)
print("\nСтатистичний опис DataFrame з NumPy масиву:")
print(df_numpy.describe())

# 2.3 Індексація (iloc, loc)

# Використання iloc (індексація по числових індексах)
print("\nВибір першого рядка за допомогою iloc:")
print(df_dict.iloc[0])

# Використання loc (індексація за мітками або умовами)
print("\nВибір квітів, де Months of flowering > 4 за допомогою loc:")
print(df_dict.loc[df_dict['Months of flowering'] > 4])
```

```python
#Part 3
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error

# 1. Генерація даних для лінійної функції з похибками вимірювань
np.random.seed(0)
x = np.linspace(0, 10, 50)  
y_true = 2 * x + 1  
noise = np.random.normal(0, 1, size=x.shape)  
y_noisy = y_true + noise  

# 2. Побудова графіку
plt.figure(figsize=(10, 6))
plt.plot(x, y_true, label='True Function (y = 2x + 1)', color='blue')
plt.scatter(x, y_noisy, label='Noisy Data', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Лінійна функція з похибками вимірювань')
plt.legend()
plt.grid(True)
plt.show()

# 3. Обчислення метрик похибок
mae = mean_absolute_error(y_true, y_noisy)  # MAE
mse = mean_squared_error(y_true, y_noisy)  # MSE

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")

# 4. Запис результатів у CSV файл
results = {
    'x': x,
    'y_true': y_true,
    'y_noisy': y_noisy,
    'MAE': [mae] * len(x),  # Повторюємо MAE для кожного x
    'MSE': [mse] * len(x)   # Повторюємо MSE для кожного x
}

df_results = pd.DataFrame(results)
df_results.to_csv('linear_function_with_errors.csv', index=False)

print("\nРезультати збережені у файл 'linear_function_with_errors.csv'.")
```
